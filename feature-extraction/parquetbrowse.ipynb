```python
import pandas as pd

def read_lines_of_parquet(file_path: str, n: int = 5, random: bool = False):
    """
    Reads either the first 'n' lines or 'n' random lines of a Parquet file
    and prints out the data in a dictionary format.
    
    :param file_path: Path to the Parquet file.
    :param n: Number of lines to read. Defaults to 5.
    :param random: Boolean flag to read random lines if set to True. Defaults to False.
    """
    # Read the entire Parquet file
    df = pd.read_parquet(file_path, engine='pyarrow', columns=None)
    
    # Select the rows based on the 'random' parameter
    if random:
        df_sample = df.sample(n=n)
    else:
        df_sample = df.head(n)
    
    # Initialize the dictionary to store the data
    data = {column: df_sample[column].tolist() for column in df_sample.columns}
    
    # Print the data in the required format
    print("data = {")
    for column, values in data.items():
        print(f'    "{column}": {values},')
    print("}")


read_lines_of_parquet('floor/sample.parquet', n=5, random=True)
```
